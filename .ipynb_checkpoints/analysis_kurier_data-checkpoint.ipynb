```python
def one_word_token(row):
    return [token for token in tokenize.word_tokenize(row) if token.lower() not in stopwords]

def two_words_token(row, n=2):
    import re
    tokens = []
    items = row.split(' ')
    items = [i for i in items if i.lower() not in stopwords]
    # Removes extra spaces in the middle of the string
    row = ' '.join([i.strip() for i in items if i])
    
    while row:
        m = re.match(r'^((?:[^ ]* ){%d}[^ ]*) (.*)' % (n-1), row.strip())
        if m:
            groups = m.groups()
            tokens.append(groups[0])
            row = groups[1]
        else:
            break
    return tokens

def three_words_token(row):
    return two_words_token(row, n=3)
```

```python
def get_stemmed_words(data_frame, my_tokenizer=one_word_token):
    stemmed_words = []
    stemmer = nltk.stem.RSLPStemmer()

    # Didn't filter by rows classified as 1
    for i, conteudo in data_frame.iterrows():
        row = conteudo['conteudo']
        words = my_tokenizer(row)
        for word in words:
            if not word:
                continue
            if word.isdigit():
                continue

            stemmed_words.append(stemmer.stem(word))
    return stemmed_words
```

## Most frequent words with stemming and removing stopwords

```python
from collections import Counter
stemmed_words = get_stemmed_words(df_vara)
c = Counter(stemmed_words)
c.most_common(50)
```

## Analysing phrases with two tokens, stemming and removing stopwords

```python
from collections import Counter
stemmed_words = get_stemmed_words(df_vara, my_tokenizer=two_words_token)
c = Counter(stemmed_words)
c.most_common(50)
```

## Analysing phrases with three tokens, stemming and removing stopwords

```python
from collections import Counter
stemmed_words = get_stemmed_words(df_vara, my_tokenizer=three_words_token)
c = Counter(stemmed_words)
c.most_common(50)
```
