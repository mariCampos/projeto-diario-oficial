# Descriptive Analysis

It is a simple analysis from data

```python
import json
import numpy
import matplotlib.pyplot as plt
```

```python
with open('../data/textos_dia_mes.json') as json_file:
    data = json.load(json_file)
data = data['data']
```

```python
names = []
values = []
for index,item in enumerate(data):
    names.append(item['day_month'])
    values.append(item['qtd'])

fig, axs = plt.subplots()
axs.bar(names, values)
fig.suptitle('Palavras por dia em um mês')
```

```python
with open('../data/textos_dia_semana.json') as json_file:
    data = json.load(json_file)
data = data['data']
```

```python
names = []
values = []
for index,item in enumerate(data):
    names.append(item['day_week'])
    values.append(item['qtd'])

fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('Palavras por dia em uma semana')
```

```python
fig, ax = plt.subplots()
ax.barh(names, values)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlabel='Quantidade', ylabel='Dias na semana',
       title='Palavras por dia em uma semana')
```
